```python
import bydoux_tools as bt


import pandas as pd
import requests

url = "https://en.wikipedia.org/wiki/List_of_S%26P_500_companies"

# Tlcharger la page avec un User-Agent
headers = {"User-Agent": "Mozilla/5.0 (X11; Linux x86_64)"}
resp = requests.get(url, headers=headers)
resp.raise_for_status()  # pour lever une erreur si 403 ou 404

# Lire les tables  partir du HTML
tables = pd.read_html(resp.text)

# Trouver la table contenant "Symbol"
df = None
for t in tables:
    if "Symbol" in t.columns:
        df = t
        break

if df is None:
    raise RuntimeError("Impossible de trouver la table des tickers.")

tickers = df["Symbol"].astype(str).str.strip().tolist()

print("tickers =", tickers)


#tickers = bt.get_tsx()

for ticker in tickers[::-1]:
    print(ticker)
    try:
        tbl = bt.read_quotes(ticker)
        if tbl is None:
            continue
    except Exception as e:
        print(f"Error processing {ticker}: {e}")
```
